```python
# imports

import pandas as pd
```

```python
# load files

cases_df = pd.read_csv('covid-19-all.csv',
                       header=0,
                       names=['country',
                              'state',
                              'lat',
                              'long',
                              'confirmed',
                              'recovered',
                              'dead',
                              'date'])
map_df = pd.read_csv('geoMap.csv',
                     header=0,
                     names=['country',
                            'mask',
                            'quarantine',
                            'symptom',
                            'trump'])
timeline_df = pd.read_csv('multiTimeline.csv',
                          header=0,
                          names=['date',
                                 'mask',
                                 'quarantine',
                                 'symptom',
                                 'trump'])
```

```python
# merge and normalize timeline and cases dataframes

df = cases_df.groupby('date').sum().merge(timeline_df, on='date')

df.confirmed = df.confirmed.divide(df.confirmed.max() / 100)
df.recovered = df.recovered.divide(df.recovered.max() / 100)
df.dead = df.dead.divide(df.dead.max() / 100)
df.tail()
```

```python
# write timeline csv

df.to_csv('timeline.csv', index=False)
```

```python
# merge and normalize map and cases dataframes
# this data is only for 3-15

map_df.at[36, 'country'] = 'US'
df = cases_df[cases_df.date=='2020-03-15'].groupby('country').sum().merge(map_df, on='country')
df['normalized confirmed'] = df.confirmed.divide(df.confirmed.max() / 100)
df = df.fillna(0)
```

```python
# write map csv

df.to_csv('map.csv', index=False)
```
